```python
import numpy as np
from sklearn.neighbors import KernelDensity
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Generate some random data
np.random.seed(0)
data = np.concatenate((np.random.normal(-2, 1, 500), np.random.normal(2, 1, 500)))

# Perform KDE estimation
kde = KernelDensity(bandwidth=0.8).fit(data[:, np.newaxis])
kde_scores = kde.score_samples(data[:, np.newaxis])

# Perform K-means clustering on KDE scores
kmeans = KMeans(n_clusters=2, random_state=0).fit(kde_scores.reshape(-1, 1))
labels = kmeans.labels_

# Plot the original data and cluster assignments
plt.figure(figsize=(8, 6))
plt.scatter(data, np.zeros_like(data), c=labels, cmap='viridis')
plt.title('1D Clustering with KDE')
plt.xlabel('Data')
plt.yticks([])
plt.show()
```

```python
# Generate some random data with two groups
np.random.seed(0)
data = np.concatenate([np.random.normal(0, 1, 1000), np.random.normal(5, 1, 1000)])[:, np.newaxis]
```

```python
# Perform KDE clustering
kde = KernelDensity(kernel='gaussian', bandwidth=0.75).fit(data)
```

```python
kde_scores = kde.score_samples(data)
```

```python
# Perform K-means clustering on the KDE scores
kmeans = KMeans(n_clusters=2, random_state=0).fit(kde_scores.reshape(-1, 1))
labels = kmeans.labels_
```

```python
# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(data, np.zeros_like(data), c=labels, cmap='viridis', s=50)
plt.title('1D Clustering using KDE', fontsize=14)
plt.xlabel('Data Points', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks([])
plt.show()
```
